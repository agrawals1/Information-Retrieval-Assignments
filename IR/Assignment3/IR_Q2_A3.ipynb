```python
import numpy as np
import networkx as nx
```

```python
#get all nodes using random walk function read_edgelist
DiGraph = nx.read_edgelist(r'/content/wiki-Vote.txt', comments='#', delimiter='\t', create_using=nx.DiGraph) #DiGraph used for directed graphs
```

##1. PageRank score for each node

---



```python
#finding page rank for the graph with default parameters
pageRankScores = nx.pagerank(DiGraph) #default: alpha=0.85 and max_iter=100
pageRankScores
# sorting pages based on high to low score
dict(sorted(pageRankScores.items(), key=lambda item: item[1], reverse = True))
```

##2. Authority and Hub score for each node

##2.1 Hub score for each node

```python
#Finding Authority and Hub scores together
hubs, authorities = nx.hits(DiGraph, max_iter = 100, normalized = True)
#arranged the hub score in decreasing order
dict(sorted(hubs.items(), key=lambda item: item[1], reverse = True))
```

##2.2 Authority Score for all nodes

```python
#sorting authorities score using dictionary sort in reverse order
dict(sorted(authorities.items(), key=lambda item: item[1], reverse = True))
```
